```python
review_content = """# HardwareX Manuscript Review: Induction Furnace Retrofit

This is a peer review of the draft manuscript "Retrofitting a commercial RF induction generator into a computer-controlled, vacuum-integrated annealing system for reactive-metal grain growth." The manuscript targets the journal *HardwareX*. Below is detailed, actionable feedback assessing the manuscript against *HardwareX* author guidelines, reviewing the "retrofit" framing, assessing the bill of materials, guiding the validation section, and providing a final submission checklist.

---

## 1. Completeness and Quality Against HardwareX Required Sections

A comparison of the current draft (`paper.md`) against the official [HardwareX Article Template](https://zenodo.org/records/5078227) and Guide for Authors reveals several gaps:

*   **Specifications Table:** 
    *   **Structure:** *HardwareX* requires specific, strictly-formatted rows for the Specifications Table: *Hardware name*, *Subject area* (from a set list), *Hardware type* (from a set list), *Closest commercial analog*, *Open source license*, *Cost of hardware*, and *Source file repository*. 
    *   **Missing Fields:** The draft is missing the mandatory **"Source file repository"** row (which must point to a permanent DOI, e.g., Zenodo/OSF, not just a GitHub link). Furthermore, the draft includes extra non-standard rows (Induction generator, Power-control input, Vacuum level, etc.). These technical specs should be moved to the *Hardware Description* section, or kept strictly as optional rows at the bottom of the table.
    *   **OSHWA Certification (Optional but recommended):** Add a row for the OSHWA Certification UID if the project obtains one.
*   **Hardware in Context:** This section is well-written. It clearly explains why a lab would retrofit an existing generator rather than buy a commercial vacuum furnace.
*   **Hardware Description:** Strong, but must refer to the design files more explicitly. Make sure to describe the software/firmware role explicitly, referencing the LabVIEW and MATLAB code.
*   **Design Files Summary:** 
    *   *HardwareX* requires four columns: *Design filename*, *File type*, *Open source license*, and *Location of the file*. The draft has this format.
    *   **Missing:** The "Location" column must ultimately contain DOIs to a permanent repository (e.g., Zenodo), not relative paths (e.g., `../code/`).
*   **Bill of Materials (BOM):**
    *   The BOM in the draft needs to exactly match the *HardwareX* template columns: *Designator, Component, Number, Cost per unit - currency, Total cost - currency, Source of materials, Material type*.
    *   The draft currently splits the BOM into three tables. *HardwareX* strictly uses a single table format. You should merge them into one table and use the *Designator* column (e.g., GEN, CTRL, RETRO, CONS) or the *Component* column to visually group the retrofit vs. the generator.
*   **Build Instructions:** The text is a good start but lacks visual aids. *HardwareX* heavily emphasizes visual instructions. The draft states `[TODO: curate a clean, photographed build sequence]`. This is critical. You cannot publish the paper without step-by-step photos of the coil winding, vacuum assembly, and control wiring. 
*   **Operation Instructions:** Condensing the SOP is correct. Ensure safety hazards are explicitly highlighted here as well as in the Safety section.
*   **Validation & Characterization:** (See section 4 below).
*   **Safety:** Excellent. Good inclusion of lockout-tagout and vacuum turbo-pump warnings.
*   **Declarations:** The draft correctly notes CRediT roles, Acknowledgements, and Competing Interests are needed. A formal "Ethics statements" section (stating no human/animal subjects were used) is technically required by the template before the CRediT statement.

---

## 2. The "Retrofit/Modernization" Framing and Generator-Agnostic Argument

**The framing is highly convincing.** Repurposing existing induction heads by wrapping them in modern digital PID control, high-vacuum chambers, and custom susceptors is a highly valuable, accessible approach for academic labs.

**How to strengthen it:**
*   **Clarify the Control Interface:** The core of the "generator-agnostic" argument rests on the induction heater possessing an analog power input (4-20 mA or 0-5V). The manuscript must explicitly state how a user can determine if their existing generator supports this. Specify the exact signal requirements the LabVIEW DAQ outputs.
*   **The CYSI GP-15A Narrative:** The history regarding the Chinese CYSI GP-15A unit is useful context but currently reads slightly disjointed. In the *Hardware in context* section, keep the focus on the successful USA-sourced CEIA unit and the LEPEL prototype. Relegate the CYSI failure to a brief footnote or omit it, as it dilutes the reproducible build instructions.
*   **Susceptor Generality:** The custom graphite susceptor is a major contribution because it allows RF heating of *non-conductive* samples (like YSZ). Emphasize this capability more in the abstract and Introduction—it makes the hardware universally applicable to ceramics and polymers, not just reactive metals.

---

## 3. Bill of Materials Completeness and Reproducibility

The BOM (`docs/induction_parts_list.xlsx`) is quite good for the retrofit parts, but currently relies on `[TODO]` for the heavy equipment.

**Essential to fill in:**
1.  **CEIA Power Cube 6 kW & Master Controller v3+ Pricing:** You must provide the approximate quote/cost for the CEIA generator and controller. Readers need to know if the "base" generator costs $5k or $30k to understand the true cost relative to a turn-key $50k system.
2.  **Chiller Pricing:** Recirculating chillers are expensive. Specify the model and cost (or note "Facility water" if a chiller is not strictly required).
3.  **National Instruments DAQ:** The exact NI DAQ model (e.g., USB-6009) and cost must be specified. Generic "LabVIEW-compatible DAQ" is not reproducible enough. 
4.  **Formatting:** Merge the three BOM blocks into the single 7-column table required by *HardwareX*.
5.  **Currency and Date:** Add "(USD, 2024)" to the cost column headers.

---

## 4. Validation: Making it Publishable

The proposed validation plan (Power-temp calibration, ramp/soak profile, and SEM/EBSD microstructure) is excellent and aligns perfectly with *HardwareX* expectations for demonstrating operational utility.

**To make the validation publishable, you must include:**
1.  **Figure 1: System Performance (Thermal).** Plot the Power-setpoint (mA) vs. Sample Temperature (°C) calibration curve. Overlay this with a plot showing the *stability* of the temperature at a soak setpoint (e.g., 1200 °C over 2 hours) to prove the PID closed-loop control works. Plotting open-loop (drifting) vs. closed-loop (stable) on the same graph is a highly convincing validation.
2.  **Figure 2: Microstructural Validation (SEM/EBSD).** 
    *   Show Side-by-Side: As-received Ni/Fe vs. Annealed Ni/Fe.
    *   Include a scale bar on all SEM images.
    *   Include a simple histogram of grain size distribution (from EBSD data) demonstrating the successful grain growth.
3.  **Figure 3: System Overview (Photographic).** The schematics are good, but a high-quality photograph of the actual completed system, fully assembled on the cart with the glowing quartz tube and pyrometer, is mandatory for validation.
4.  **Address the Missing Data Logs:** The text claims 100+ runs in `docs/data_log/`, but the provided workspace only contains the `extracted-context/`. Ensure the actual raw data logs (.lvm, .xlsx) and raw micrographs are uploaded to the final repository so reviewers can verify them.

---

## 5. Highest-Priority Gaps & Submission Checklist

### Critical HardwareX Policy Gaps
1.  **Missing Permanent Repository (DOI):** GitHub is not sufficient. You must create a release on Zenodo, Mendeley Data, or OSF, mint a DOI, and put that DOI in the *Specifications Table* and *Design Files Summary*.
2.  **LabVIEW Proprietary Format:** *HardwareX* requires open/editable formats. Since LabVIEW `.vi` files require proprietary software to read, you **must** export the block diagrams to PDF or PNG and include them in the design files repository so the control logic is accessible to all.
3.  **Missing Hardware License:** You must select and apply an OSHWA-approved license (e.g., CERN-OHL-S v2) to the hardware/CAD files, and an OSI-approved license (e.g., MIT, GPL) to the LabVIEW/MATLAB code. Add a `LICENSE` file to the root repository and update the *Specifications Table*.
4.  **CAD Formats:** The graphite crucible requires a machining drawing. Provide this as an open 3D format (STEP, FreeCAD) or standard 2D PDF drawing. 

### Ordered Checklist Before Submission
1.  [ ] **Acquire Missing Data:** Pull the CEIA Generator, Controller, Chiller, and NI DAQ model numbers and prices into the BOM.
2.  [ ] **Format Tables:** Merge the BOM into a single 7-column table. Remove non-standard rows from the Specifications Table and add "Source file repository" row.
3.  [ ] **Generate Figures:** 
    *   Use `plotheatcurve.m` to generate the PID stability / thermal ramping graph.
    *   Format the SEM/EBSD images of grain growth.
    *   (Review of the provided coil images: Figure `08`, `10`, and `12` are clean and publishable. Figures `03`, `04`, and `05` are informal workshop photos and should not be included without major cleanup).
4.  [ ] **Export Code:** Export all LabVIEW `.vi` block diagrams to PDF/PNG.
5.  [ ] **Add Licenses:** Apply CERN-OHL-S to hardware files and MIT to software files.
6.  [ ] **Create Zenodo Release:** Upload the repository, generate a DOI, and update all `[TODO]` links in the manuscript.
7.  [ ] **Final Polish:** Add an Ethics Statement, CRediT authorship statement, and verify all visual build instructions are clear.
"""
with open(review_path, "w") as f:
    f.write(review_content)
```
