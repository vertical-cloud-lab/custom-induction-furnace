```python
import numpy as np

def control_metrics(csv_path, setpoint, soak_pad_s=120, deadband_C=10):
    d = pd.read_csv(csv_path)
    t = d['elapsed_s'].values
    T = d['temperature_C'].values
    P = d['power_mA'].values
    # Define "soak" by temperature within setpoint band
    in_band = np.abs(T - setpoint) <= deadband_C
    if not in_band.any():
        return None
    # First crossing of (setpoint - deadband) -> rise; first time entering band -> settle time
    idx_band = np.where(in_band)[0]
    soak_start = idx_band[0]
    soak_end = idx_band[-1]
    # Rise time: 10% to 90% from start to soak setpoint
    T0 = T[0]
    span = setpoint - T0
    t10 = t[np.argmax(T >= T0 + 0.1*span)] if (T >= T0 + 0.1*span).any() else np.nan
    t90 = t[np.argmax(T >= T0 + 0.9*span)] if (T >= T0 + 0.9*span).any() else np.nan
    rise = t90 - t10
    # Overshoot
    peak = T[:soak_end+1].max()
    overshoot = peak - setpoint
    # Settling time: time from t10 until T stays within deadband for >= soak_pad_s
    settle_t = np.nan
    for i in range(soak_start, len(T)):
        # check if from here onward stays in band for soak_pad_s
        window_end = t[i] + soak_pad_s
        j = np.searchsorted(t, window_end)
        if j > i and np.all(in_band[i:j]):
            settle_t = t[i] - t10
            break
    # Soak window: where in band continuously after settle
    soak_T = T[soak_start:soak_end+1][in_band[soak_start:soak_end+1]]
    soak_mean = soak_T.mean()
    soak_sd = soak_T.std(ddof=1)
    max_dev = np.max(np.abs(soak_T - setpoint))
    soak_dur_min = (t[soak_end] - t[soak_start]) / 60.0
    # Linear drift across soak (degrees per hour)
    soak_t_in = t[soak_start:soak_end+1]
    if len(soak_t_in) > 5:
        coef = np.polyfit(soak_t_in, T[soak_start:soak_end+1], 1)
        drift_per_hr = coef[0] * 3600
    else:
        drift_per_hr = np.nan
    return {
        'setpoint_C': setpoint,
        'rise_10_90_s': rise,
        'overshoot_C': overshoot,
        'settle_to_band_s': settle_t,
        'soak_mean_C': soak_mean,
        'soak_sd_C': soak_sd,
        'soak_max_dev_C': max_dev,
        'soak_duration_min': soak_dur_min,
        'soak_drift_C_per_hr': drift_per_hr,
        'soak_n_points': len(soak_T),
    }

# Compute for all available per-run CSVs using soak_temp_C from run_summary
results = []
df['run_key'] = df['run'].str.replace(r' \(.*\)$', '', regex=True)
for f in csvs:
    run_id = f.replace('.csv','')
    row = df[df['run'] == run_id]
    if row.empty:
        continue
    sp = row['soak_temp_C'].values[0]
    # Use the nominal setpoint from filename if available
    # parse e.g. "1400C" => 1400
    import re
    m = re.search(r'_(\d{3,4})C_', f)
    sp_nom = int(m.group(1)) if m else round(sp)
    metrics = control_metrics(f'{wd}/run-plots/{f}', sp_nom)
    metrics['run'] = run_id
    metrics['nominal_setpoint_C'] = sp_nom
    results.append(metrics)

m_df = pd.DataFrame(results)
print(m_df.to_string())
```

```python
# Power -> steady-state-T calibration: pull soak-mean power vs soak-mean T for each run
def soak_power(csv_path, setpoint, deadband_C=10):
    d = pd.read_csv(csv_path)
    in_band = np.abs(d['temperature_C'] - setpoint) <= deadband_C
    if not in_band.any():
        return None
    sub = d[in_band]
    return {'soak_mean_T': sub['temperature_C'].mean(),
            'soak_mean_P_mA': sub['power_mA'].mean(),
            'soak_sd_P_mA': sub['power_mA'].std(ddof=1),
            'n': len(sub)}

cal = []
for f in csvs:
    run_id = f.replace('.csv','')
    import re
    m = re.search(r'_(\d{3,4})C_', f)
    if not m:
        continue
    sp = int(m.group(1))
    res = soak_power(f'{wd}/run-plots/{f}', sp)
    if res:
        res['run'] = run_id
        res['setpoint_C'] = sp
        # detect material
        res['material'] = 'Ni200' if 'Ni200' in run_id else 'Ni4N5'
        res['has_flow'] = df[df['run']==run_id]['has_flow'].values[0]
        cal.append(res)

cal_df = pd.DataFrame(cal)
print(cal_df.to_string())
```

```python
# Quick power-T calibration assessment - flow-yes runs only (consistent geometry)
flow_runs = cal_df[cal_df['has_flow']=='yes'].copy()
print("Flow-yes runs (consistent gas/coil/optics setup):")
print(flow_runs.to_string())
print()
# Fit simple polynomial T vs power_mA
from numpy.polynomial import polynomial as P
X = flow_runs['soak_mean_P_mA'].values
Y = flow_runs['soak_mean_T'].values
# Fit T = a + b*P
coef = np.polyfit(X, Y, 1)
pred = np.polyval(coef, X)
resid = Y - pred
print(f"Linear fit T = {coef[1]:.1f} + {coef[0]:.1f}*P_mA")
print(f"R^2 = {1 - resid.var()/Y.var():.3f}, residual SD = {resid.std(ddof=2):.1f} °C, n = {len(X)}")
print()
print("Non-flow runs (different optical/setup era — keep separate):")
print(cal_df[cal_df['has_flow']=='no'].to_string())
```

```python
# The flow-yes group has Ni200 at lower P than Ni4N5 - that's because Ni4N5 vs Ni200 are different alloys/geometries
# Honest assessment: only the no-flow Ni4N5 set spans 4 setpoints with same setup
nf = cal_df[(cal_df['has_flow']=='no') & (cal_df['material']=='Ni4N5')]
X = nf['soak_mean_P_mA'].values; Y = nf['soak_mean_T'].values
# But IFrun049 (1200/12h) had different soak power, suggesting different load or geometry vs IFrun039 (1200/6h)
# So the calibration is genuinely setup-dependent. Honest message.

print("Ni4N5 no-flow runs at varying setpoints (4 distinct setpoints, n=5 runs):")
print(nf.to_string())
print()
# Note that IFrun049 P (0.658) >> IFrun039 P (0.440) for same nominal 1200°C - this evidences the calibration depends on geometry/era
# This is exactly why the paper should NOT promise a universal power-T calibration

# Repeat fit excluding the 1200/12h IFrun049 outlier (different geometry)
nf2 = nf[~nf['run'].str.contains('IFrun049')]
X = nf2['soak_mean_P_mA'].values; Y = nf2['soak_mean_T'].values
coef = np.polyfit(X, Y, 1)
pred = np.polyval(coef, X)
print(f"Linear fit (excluding IFrun049): T = {coef[1]:.1f} + {coef[0]:.1f}*P_mA, n={len(X)}")
print(f"R^2 = {1 - (Y-pred).var()/Y.var():.4f}")
```
